```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
df
```

```python
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'])
```

```python
x = df[['Annual Income (k$)', 'Spending Score (1-100)']]
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

# Plotting the Elbow Method graph
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
km = KMeans(n_clusters=5)
km
```

```python
y_predicted = km.fit_predict(df[['Annual Income (k$)', 'Spending Score (1-100)']])
y_predicted
```

```python
df['cluster']= y_predicted
df.head()
```

```python
df0= df[df.cluster==0]
df1= df[df.cluster==1]
df2= df[df.cluster==2]
df3= df[df.cluster==3]
df4= df[df.cluster==4]
plt.scatter(df0['Annual Income (k$)'], df0['Spending Score (1-100)'], color='green')
plt.scatter(df1['Annual Income (k$)'], df1['Spending Score (1-100)'], color='red')
plt.scatter(df2['Annual Income (k$)'], df2['Spending Score (1-100)'], color='blue')
plt.scatter(df3['Annual Income (k$)'], df3['Spending Score (1-100)'], color='black')
plt.scatter(df4['Annual Income (k$)'], df4['Spending Score (1-100)'], color='orange')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
```
